```python
error=1
alp=1.5
print(x)
```

```python
import random
#w0=random.randint(1,10)
#w1=random.randint(1,10)
w0=1,w1=1
print(y)
```

```python
J=list()
J.append((0.5/51)*sum( (y-w0-(w1*x))**2 ))
print(J[0])
```

```python
iteration=0
while(error>0.00000001):
    ycap=w0+(w1*x)
    
    DJ0=(ycap-y)
    DJ1=(ycap-y)*x
    
    dw0=-alp*sum(DJ0)/51
    dw1=-alp*sum(DJ1)/51
    
    w0=w0+dw0
    w1=w1+dw1
    
    iteration+=1
    J.append((0.5/51)*sum( (y-w0-(w1*x))**2 ))
    error=abs(J[iteration]-J[iteration-1])
    print(error)
```
